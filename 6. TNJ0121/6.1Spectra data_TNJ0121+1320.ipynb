```python
### a function that calculates the continuum range
### use p = peak, it = i, do_list = doublet_list, v_all = vel_all, mlines = my_lines
def get_continuum(p, it, do_list, mlines, v_all, beg_con_r =1800 , end_con_l = -1800):
    """return multiple variables in a tuple: (beg_con_r, end_con_r, beg_con_l, end_con_l, continuum_peak_left,
    continuum_peak_left)"""
    ### continuum range right 
    #beg_con_r = 1800  # CAN BE CHANGED
    end_con_r = beg_con_r +2000
    
    vel_prevline = vel_nextline = np.array(lamb2v(mlines[it-1], p)) # in km/s
    vel_prevprevline = np.array(lamb2v(mlines[it-2], p)) # in km/s
    vel_nextline = np.array(lamb2v(mlines[it+1], p)) # in km/s
    vel_nextnextline = np.array(lamb2v(mlines[it+2], p)) # in km/s
    ### handle the close doublets which share the same continuum:
    if p in do_list:
        distance = lamb2v(mlines[it + 1], p)
        #FWHM = 1000
        beg_con_r += distance 
        #beg_con_r += (distance + FWHM) 
        end_con_r = 2000 + beg_con_r
        
    if (vel_nextline >= beg_con_r) & (vel_nextline <= end_con_r): ### take the cont. from next em. line
        FWHM = 3000  # CAN BE CHANGED
        distance = lamb2v(mlines[it+1], p) # could be initialized
        #beg_con += distance
        beg_con_r += (distance + FWHM/2)
        end_con_r = 2000 + beg_con_r
       #end_con = 1500 + distance + FWHM/2
        
    elif (vel_nextnextline >= beg_con_r) & (vel_nextnextline <= end_con_r):
        FWHM = 3000 # CAN BE CHANGED
        distance = lamb2v(mlines[i+2], p) # CAN BE CHANGED
        beg_con_r += (distance +FWHM/2)
        end_con_r = 2000 + beg_con_r
            
    condition1_r =  (vel_all >= beg_con_r) & (vel_all <= end_con_r)
    continuum_peak_right = np.extract(condition1_r, obj1.data)

    ### continuum range left
    end_con_l = -1800
    beg_con_l = end_con_l - 2000
    
        ### handle the close doublets which share the same continuum:
    if p in doublet_list:
        distance = lamb2v(mlines[it-1], p)
        #FWHM = 1000
        beg_con_l += distance 
        end_con_l = 2000 + beg_con_l
        
    if (vel_prevline >= beg_con_l) & (vel_prevline <= end_con_l): ### take the cont. from next em. line
        FWHM = 3000  # CAN BE CHANGED
        distance = lamb2v(mlines[it-1], p) # could be initialized
        beg_con_l += (distance + FWHM/2)
        end_con_l = 2000 + beg_con_l
        
    elif (vel_prevprevline >= beg_con_l) & (vel_prevprevline <= end_con_l):
        FWHM = 3000 # CAN BE CHANGED
        distance = lamb2v(mlines[it-2], p) # CAN BE CHANGED
        beg_con_l += (distance +FWHM/2)
        end_con_l = 2000 + beg_con_l
    
    condition1_l =  (vel_all >= beg_con_l) & (vel_all <= end_con_l)
    continuum_peak_left = np.extract(condition1_l, obj1.data)
    
    return beg_con_r, end_con_r, beg_con_l, end_con_l, continuum_peak_left, continuum_peak_right
```
